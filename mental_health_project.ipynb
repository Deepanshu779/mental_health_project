## 📁 Premium Multi-Page Architecture

The application is organized into a modular, highly performant multi-page layout:

```
mental_health_project/
│
├── app.py                     # Home / Landing Page & Mindfulness Center
│
├── pages/                     # Feature pages (Streamlit sidebar navigation)
│   ├── 1_Chatbot.py           # Main Companion Chatbot & Crisis Override
│   ├── 2_Mood_Tracker.py      # Mood Journey Slider & Plotly Analytics
│   ├── 3_Helpline.py          # Emergency Region Helpline Grid
│
├── utils/                     # Shared logical and styling modules
│   ├── bot_logic.py           # Unified Logical Interface & Compatibility
│   ├── bot_engine.py          # AI Gemini & Rule-based Persona Response Models
│   ├── data_store.py          # Mood Logs CSV database manager
│   └── styles.py              # CSS injects, breathing circles, & unified sidebar
│
└── mental_health_project.ipynb
```

## 🌿 1. Landing Page & Mindfulness Retreat (`app.py`)

The main entry point serves as both a welcome landing page and an interactive mindfulness retreat. It includes:
- **Rhythmic Breathing Gym:** CSS-animated breathing sphere supporting Box Breathing, 4-7-8 Deep Calm, and Resonance pacing.
- **Self-Reflection Canvas:** Prompt-based journaling space supporting custom file downloads.
- **5-4-3-2-1 Grounding Checklist:** Interactive checklist to intercept panic and anchor awareness.
- **Global State Initialization:** Prepares user states for companion chat history, active API settings, and crisis alerts.

## 💬 2. Companion Chatbot (`pages/1_Chatbot.py`)

The chatbot page facilitates comforting emotional dialogue:
- **Safety & Crisis Bounds:** Scans user entries for self-harm intent. If matched, triggers a **Crisis Override** red card that replaces the chat block and displays immediate hotlines.
- **Empathetic Personas:** Supports **Serene** (peaceful calming), **Joy** (optimistic self-compassion), and **Sage** (CBT-style cognitive reframing).
- **Google Gemini Integration:** Securely reads in-memory keys from the sidebar to activate **Gemini 1.5 Flash** for contextual, empathetic conversations. Otherwise, falls back to pre-programmed responses.
- **Auto-Mood Logging:** Scans dialogue for keywords. If a strong positive or negative tone is registered, it automatically logs a wellness rating in the background.

## 📊 3. Mood Analytics Dashboard (`pages/2_Mood_Tracker.py`)

Provides clinical-style self-reporting and visualization:
- **Reflection Form:** Slider for ratings (1-10), multiselect tags for lifestyle triggers (Sleep, Hobbies, Work, Exercise), and notes.
- **Plotly Trends:** Generates an emotional trend area chart over time.
- **Plotly Distribution:** Donut chart of primary emotional frequencies.
- **Plotly Lifestyle Correlations:** Horizontal bar chart illustrating average wellness ratings corresponding to active lifestyle triggers.

## 🚨 4. Emergency Helplines Grid (`pages/3_Helpline.py`)

A region-segmented, 24/7 confidential helpline catalog supporting the United States, United Kingdom, India, Canada, and International domains. Rendered in a high-contrast glassmorphic grid layout.

## ⚙️ 5. Shared Utilities (`utils/`)

- **`utils/styles.py`:** Standardizes visual properties (Plus Jakarta Sans font, dark background, glass-card styles, tab adjustments). Features the unified `render_sidebar()` function that manages persona selections and Gemini credentials globally across subpages.
- **`utils/bot_logic.py`:** Exposes the modular logical engine wrappers.
- **`utils/data_store.py`:** Manages the secure, local CSV-based emotional database (`mood_logs.csv`).

## 🧪 6. Compilation & Code Integrity Verification

Run the cell below to verify that all newly refactored project components are free of syntax issues and compile successfully.

````python
import os
import py_compile

def verify_project_code():
    components = [
        'app.py',
        'pages/1_Chatbot.py',
        'pages/2_Mood_Tracker.py',
        'pages/3_Helpline.py',
        'utils/bot_logic.py',
        'utils/bot_engine.py',
        'utils/styles.py',
        'utils/data_store.py'
    ]
    
    print('🔍 Verifying EmpathyBot 2.0 components...')
    success = True
    
    for filepath in components:
        if not os.path.exists(filepath):
            print(f'❌ Error: {filepath} does not exist!')
            success = False
            continue
        
        try:
            py_compile.compile(filepath, doraise=True)
            print(f'✅ Compiled successfully: {filepath}')
        except py_compile.PyCompileError as e:
            print(f'❌ Syntax Error in {filepath}: {e}')
            success = False
            
    if success:
        print('\n⭐ All EmpathyBot 2.0 components compiled flawlessly! Ready for deployment.')
    else:
        print('\n⚠️ Some components need structural review.')

verify_project_code()
````
